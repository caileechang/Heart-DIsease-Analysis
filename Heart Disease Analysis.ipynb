**Data Preprocessing**

````python
# Find missing values
# It returns a Boolean DataFrame where each element is either True or False
# Indicating whether the corresponding value in the original dataset is missing or not.
data.isnull()
````

````python
# Checking which columns are missing values
# Calculate the sum of missing values in each column of a dataset.
# It returns a Series object that displays the total number of missing values for each column.
data.isnull().sum()
````

````python
# Provides a simple check to determine if there are any missing values in the dataset and prints the appropriate message based on the result.
missing_values = data.isnull().sum()
if missing_values.sum() == 0:
    print("There are no missing values in the dataset.")
else:
    print("There are missing values in the dataset.")
````

````python
# Drop rows with missing values
data.dropna(inplace=True)
data.isnull().sum()
````

````python
# Checking if there are any duplicate rows in the dataset
data_dup = data.duplicated().any()
data_dup
````

````python
# Drop duplicate rows from the dataset and then checks if there are any remaining duplicate rows
data = data.drop_duplicates()
data_dup = data.duplicated().any()
data_dup
````

````python
# Check for inconsistent values in multiple columns
# Examine the distribution and frequency of values in each column
for col in data.columns:
    print(col)
    print(data[col].value_counts())
````

````python
# Split the features and target variable
# This type of splitting is done in machine learning tasks
# X represents the input features that will be used to make predictions, and y represents the target variable that we want to predict or model.
# This separation allows us to train machine learning models on X and evaluate their performance using y.
X = data.drop('Heart_ stroke', axis=1)
y = data['Heart_ stroke']
````

````python
print(X)
````

````python
print(y)
````

````python
# Separate categorical and continuous variables
# This separation is useful when performing different types of analysis or preprocessing on categorical and continuous variables, as they often require different treatment.
# Categorical variables may need to be encoded or transformed using techniques such as one-hot encoding, while continuous variables may require scaling or normalization.
categorical_vars = data.select_dtypes(include='object').columns
continuous_vars = data.select_dtypes(include=['int64', 'float64']).columns
````

````python
# Display the categorical variables
# The categorical_vars variable will contain the column names of the categorical variables in the dataset
print("Categorical Variables:")
print(categorical_vars)
````

````python
# Display the continuous variables
# The continuous_vars variable will contain the column names of the continuous variables.
print("\nContinuous Variables:")
print(continuous_vars)
````

````python
# Complete description of the continuous data as well as the categorical data
# Allowing to further analyze or preprocess the data based on their respective characteristics
categorical_val = []
continous_val = []
for column in data.columns:
    print("--------------------")
    print(f"{column} : {data[column].unique()}")
    if len(data[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
````

````python
# Split the data into training and testing sets
# This splitting is commonly done in machine learning tasks to evaluate the performance of a model.
# The model is trained on the training data (X_train and y_train) and then tested on the unseen testing data (X_test and y_test) to assess its generalization ability.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
train_data = X_train.join(y_train)
train_data
````

````python
# Dropping a column, replacing categorical values with numeric representations, calculating the means of certain columns, and filling missing values with the respective mean values
# Help to handle missing data
data.drop(columns = ['education'], inplace = True)
data['Gender'] = data['Gender'].replace({'Male': 0, 'Female': 1})
data['prevalentStroke'] = data['prevalentStroke'].replace({'no': 0, 'yes': 1})
data['Heart_ stroke'] = data['Heart_ stroke'].replace({'No': 0, 'yes': 1})
cigsPerDay_mean = data['cigsPerDay'].mean()
totChol_mean = data['totChol'].mean()
heartRate_mean = data['heartRate'].mean()
BMI_mean = data['BMI'].mean()
glucose_mean = data['glucose'].mean()
data['cigsPerDay'] = data['cigsPerDay'].fillna(cigsPerDay_mean)
data['totChol'] = data['totChol'].fillna(totChol_mean)
data['BMI'] = data['BMI'].fillna(BMI_mean)
data['glucose'] = data['glucose'].fillna(glucose_mean)
data['heartRate'] = data['heartRate'].fillna(heartRate_mean)
````

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# Define the column transformer for preprocessing
# Numeric columns are scaled using MinMaxScaler, and categorical columns are encoded using OneHotEncoder
# This preprocessing step is often performed before feeding the data into a machine learning model to ensure that all features are in a suitable format and have similar scales.
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), ['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']),
        ('cat', OneHotEncoder(), ['Gender', 'education', 'currentSmoker', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes'])
    ])
````

````python
# The resulting X will contain all the columns from data except for 'Heart_stroke'.
# The resulting y will be a Series containing the values of 'Heart_stroke' column.
X = data.drop(columns = ['Heart_ stroke'])
y = data['Heart_ stroke']
````

````python
# Label encoding is used to convert categorical variables into numerical representations so that they can be effectively used as inputs for machine learning algorithms
# The fit_transform() method is used on the training set to both fit the encoder on the unique categories present in the 'Gender' column and transform the column into encoded labels.
# The transform() method is then used on the testing set to transform the 'Gender' column using the fitted encoder, ensuring consistency between the training and testing data. 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X_train['Gender'] = le.fit_transform(X_train['Gender'])
X_test['Gender'] = le.transform(X_test['Gender'])
````

````python
# The fit_transform() method is used on the training set to both fit the encoder on the unique categories present in the 'education' column and transform the column into encoded labels. 
# The transform() method is then used on the testing set to transform the 'education' column using the fitted encoder, ensuring consistency between the training and testing data.
le = LabelEncoder()
X_train['education'] = le.fit_transform(X_train['education'])
X_test['education'] = le.transform(X_test['education'])
````

````python
# The fit_transform() method is used on the training set to both fit the encoder on the unique categories present in the 'prevalentStroke' column and transform the column into encoded labels.
# The transform() method is then used on the testing set to transform the 'prevalentStroke' column using the fitted encoder, ensuring consistency between the training and testing data.
le = LabelEncoder()
X_train['prevalentStroke'] = le.fit_transform(X_train['prevalentStroke'])
X_test['prevalentStroke'] = le.transform(X_test['prevalentStroke'])
````

````python
# Performs label encoding on the target variable 'y_train'
# The target variable 'y_train' is encoded using the LabelEncoder object le. 
# The fit_transform() method is used to both fit the encoder on the unique categories present in 'y_train' and transform the variable into encoded labels.
le = LabelEncoder()
y_train = le.fit_transform(y_train)
````

````python
# Performs feature scaling using the StandardScaler on the training and testing sets
# Feature scaling is a common preprocessing step in machine learning that helps bring features to a similar scale
# Can improve the performance and convergence of many machine learning algorithms
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

Regression Model Development  

````python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
rf = RandomForestClassifier()
xg = XGBClassifier()
rf.fit(X_train, y_train)
xg.fit(X_train, y_train)
````

````python
y_pred_rf = rf.predict(X_test)
y_pred_xg = xg.predict(X_test)
````

````python
# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
````

````python
# Make predictions on the test set
y_pred = model.predict(X_test_scaled)
````

Model Evaluation

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_test_int = [1 if label == 'yes' else 0 for label in y_test]
y_pred = model.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test_int, y_pred)
print('Accuracy:', accuracy)

# Calculate recall
recall = recall_score(y_test_int, y_pred)
print('Recall:', recall)

# Calculate F1 score
f1 = f1_score(y_test_int, y_pred)
print('F1 Score:', f1)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve


# calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test_int, y_pred)
plt.plot(fpr, tpr)
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

# calculate EROC
cm = confusion_matrix(y_test_int, y_pred >= 0.5)
eroc = cm[0, 1] / (cm[0, 1] + cm[1, 0])
print('EROC: %.3f' % eroc)

# calculate MER
n = len(y_test_int)
n_errors = sum(abs(y_test_int - (y_pred >= 0.5)))
mer = n_errors / n
print('MER: %.3f' % mer)
````
